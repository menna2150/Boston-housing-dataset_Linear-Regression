#Predictive Analysis of Boston Housing Data#
The Boston Housing Dataset is a classical benchmark in regression analysis and machine learning, widely used for modeling and evaluating techniques that predict continuous variables. This dataset captures a diverse set of socioeconomic and structural attributes of residential neighborhoods in the Boston area. The primary objective of this analysis is to predict the median value of owner-occupied homes (MEDV), which serves as the target variable.The study begins with a comprehensive Exploratory Data Analysis (EDA) phase to uncover variable distributions, detect outliers, and examine inter-variable relationships through histograms and correlation matrices. Recognizing the importance of data quality, particular emphasis is placed on handling missing and infinite values, which are imputed using the K-Nearest Neighbors (KNN) imputer for a robust preprocessing pipeline.Building upon these insights, the dataset is modeled using two distinct regression techniques: Linear Regression and Decision Tree Regression. These models are implemented and evaluated to compare their predictive capabilities, interpretability, and sensitivity to data characteristics. The results not only inform the efficacy of each approach but also provide a foundation for more advanced predictive modeling strategies.



#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

#Read Data

```python
df=pd.read_csv("/content/HousingData.csv")
```

#preprocessing

```python
df=df.dropna()
```

```python
x=df.drop('MEDV',axis=1)
```

```python
y=df['MEDV']
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
```
